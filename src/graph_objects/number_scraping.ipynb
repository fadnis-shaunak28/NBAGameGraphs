```python
def getScoreBoardGames():
    scoreboard = nba_live.ScoreBoard()
    games = scoreboard.games.get_dict()

    # get game Id and the teams playing
    games_simplified = []
    for game in games:
        gid = game['gameId']
        home_team = game["homeTeam"].get("id", 0)
        away_team = game["awayTeam"].get("teamName", 0)
        game_slug = [gid, home_team, away_team]
        games_simplified.append(game_slug)
    
    return games_simplified
```
